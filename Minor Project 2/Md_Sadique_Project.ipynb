## Phase 3: Outlier Analysis

#### Step 1: Numerical Columns Select Karo

```python
num_cols = df.select_dtypes(include=np.number).columns
num_cols
```

#### Step 2: Check Skewness

```python
df[num_cols].skew()
```

#### Step 3: Draw Box Plot

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

#### Step 4: Detect Outliers Using IQR

```python
for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} : {len(outliers)} Outliers")
```

```python
df.head()
```

#### Step 5: Handle Outliers (Trimming)

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

#### Step 6: Transformation (Log/Sqrt Scaling)

```python
df[col] = np.log1p(df[col])
```

## Phase 4: Bivariate & Multivariate Exploration

#### Step 1: Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.show()
```

#### Step 2: Explain Strong Positive and Negative Relationships

Based on the correlation heatmap:

There is a strong positive correlation between Price and RAM, which indicates that laptops with higher RAM generally have higher prices.
There is a weak positive correlation between Price and Weight, which means laptop weight has only a small influence on its price.

Note: These observations should be written according to the results shown in your correlation heatmap.

#### Step 3: Box Plot (Categorical vs Numerical)

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Company', y='Price_euros', data=df)
plt.xticks(rotation=90)
plt.show()
```

#### Step 4: Bar Chart (Categorical vs Numerical)

```python
df.groupby("Company")["Price_euros"].mean().plot(kind="bar")

plt.xlabel("Company")
plt.ylabel("Average Price (Euros)")
plt.title("Average Laptop Price by Company")
plt.xticks(rotation=45)

plt.show()
```

#### Step 5: Correlation ≠ Causation

#### Observed correlation does not necessarily imply direct causation. Other hidden factors may also influence the relationship between variables.

## Phase 5: Feature Engineering & Encoding Preparation

#### Step 1: Create a New Feature

```python
df["Storage_Type"] = df["Memory"].apply(lambda x: "SSD" if "SSD" in x else "HDD")
```

```python
df[["Memory","Storage_Type"]].head()
```

#### Step 2: Identify Ordinal and Nominal Columns

##### Ordinal (Ranked)Columns:
Ram (4GB < 8GB < 16GB < 32GB)
Inches (Screen size can be treated as ordered)

##### Nominal (Unranked)Columns:
Company
Product
TypeName
ScreenResolution
Cpu
Memory
Gpu
OpSys


#### Step 3: Label Encoding (Ordinal)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Storage_Type"] = le.fit_transform(df["Storage_Type"])
```

#### Step 4: One-Hot Encoding (Nominal)

```python
df = pd.get_dummies(
    df,
    columns=[
        "Company",
        "Product",
        "TypeName",
        "ScreenResolution",
        "Cpu",
        "Memory",
        "Gpu",
        "OpSys"
    ],
    drop_first=True
)
```
